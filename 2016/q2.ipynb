```python
'''
Infinite squares
- 4 or more people, each move to adjacent square
- Each step, a new person is added
- Keep moving until no overcrowded
- People spawned in the 5*5 square and nowhere else
- starting position and added modifiers 
'''
import numpy as np

def adjacent_squares(coord):
    x = coord[0]
    y = coord[1]
    return [[x-1,y],
            [x,y+1],
            [x+1,y],
            [x,y-1]]


class Landscape():
    def __init__(self,position,modifiers,steps):
        self.position = position
        self.steps = steps
        self.modifiers = modifiers
        self.grid = np.zeros((15,15))
        self.indices = []
        for a in range(15):
            for b in range(15):
                self.indices.append([a,b])
                
        self.five_by_five = []
        for a in range(5):
            for b in range(5):
                self.five_by_five.append(np.array([a,b]) + [4,4])
                
    def update(self):
        while True:
            overcrowded = np.transpose(np.where(self.grid == 4))
            if not(list(overcrowded)):
                break
            print(overcrowded)
            for point in list(overcrowded):
                self.grid[point[0]][point[1]] = 0
                print(point)
                adjacent_points = adjacent_squares(point)
                for point in adjacent_points:
                    if list(point) in list(self.indices):
                        print("YO")
                        self.grid[point[0]][point[1]] += 1
                    
    def add_people(self):
        place_coord = self.five_by_five[self.position-1]
        self.grid[place_coord[0]][place_coord[1]] += 1
        print(self.position)
        self.position += self.modifiers[0]
        if self.position > 25:
            self.position -= 25
        self.modifiers.append(self.modifiers[0])
        self.modifiers.pop(0)
    
    def execute(self):
        for i in range(self.steps):
            self.add_people()
            #print(landscape.grid)
            #print()
            self.update()
            #print(landscape.grid)
            #print("DASDF")
            
    def print_grid(self):
        five_by_five = np.array(self.five_by_five).reshape(5,5,2)

        for row in five_by_five:
            row_string = ''
            for value in row:
                row_string += str(int(self.grid[value[0]][value[1]]))
                row_string += ' '
            print(row_string)
```

```python
landscape = Landscape(3,[2,3,5,7,11,13],150)
landscape.execute()
landscape.print_grid()
```
