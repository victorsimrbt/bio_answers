```python
lucky = []
odd_numbers = [i for i in range(10100) if i % 2 != 0]

def update(n):
    if n == max(odd_numbers):
        return None
    #print(odd_numbers,n)
    lucky.append(n)
    idx = odd_numbers.index(n)
    new_lucky = odd_numbers[idx+1]
    numbers = [odd_numbers[i] for i in range(len(odd_numbers)) if (i+1) % new_lucky == 0 and i != 0]
    for num in numbers:
        odd_numbers.remove(num)
    update(new_lucky)
    
num = int(input())
update(1)
lucky_copy = lucky[:]
lucky_copy.append(num)
lucky_copy.sort()

num_index = lucky_copy.index(num)
print(lucky_copy[num_index-1],lucky_copy[num_index+1])
```
