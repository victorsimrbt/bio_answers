```python
''' 
Rules of Note:

- 10 by 10 grid
- Ships can only be placed vertically or horizontally
- Bottom left is (0,0)
- A,C,M,R

Steps:
r = (a*r+c) % m 
Unit digits of r as x, Tens as y coordinate
r = (a*r+c) % m 
If valid, if r is even, horizontal. If odd, vertically upwards
If ship placed, stop. Else, restart
'''
def check_adjacent(point):
    x,y = point
    adjacent_points = [[x-1,y-1],
                       [x-1,y],
                       [x-1,y+1],
                       [x,y+1],
                       [x+1,y+1],
                       [x+1,y],
                       [x+1,y-1]]
    return adjacent_points

import numpy as np
class Grid:
    def __init__(self):
        self.ships = []
        self.grid = np.zeros((10,10))
        self.indices = []
        for a in range(10):
            for b in range(10):
                self.indices.append([a,b])
    def valid(self,coords):
        # ! coords is a list of coordinations of the ship
        for coord in coords:
            adjacent_points = check_adjacent(coord)
            adjacent_points.append(coord)
            for point in adjacent_points:
                if point in self.indices:
                    if self.grid[point[0]][point[1]] == 0:
                        pass
                    else:
                        return False
        return True
    def place_ship(self,coords,orientation):
        for coord in coords:
            self.grid[coord[0]][coord[1]] = 1
        self.ships.append([coords[0][0],coords[0][1],orientation])
        
class Player:
    def __init__(self,a,c,m):
        self.a = a
        self.c = c
        self.m = m
        self.r = 0
        self.ship_val = [4,3,3,2,2,2,1,1,1,1]
    def ship_points(self,coord):
        points = []
        orientation = [np.array([0,1]) if self.r % 2 == 0 else np.array([1,0])][0]

        last_point = coord
        for i in range(self.ship_val[0]):
            new_point = list((orientation*i + np.array(last_point)))
            points.append(new_point)
        return points
    def place_ship(self,grid):
        while True:
            print("YEAH")
            self.r = (self.a*self.r+self.c) % self.m 
            print("R",self.r)
            unit = int(str(self.r)[-1])
            if len(str(self.r)) == 1:
                tens = 0
            else:
                tens = int(str(self.r)[-2])
            coord = [unit,tens]
            print(coord)
            self.r = (self.a*self.r+self.c) % self.m 
            print("R",self.r)
            orientation = ["H" if self.r % 2 == 0 else "V"]
            coords = self.ship_points(coord)
            print(coords)
            if grid.valid(coords):
                print("YES")
                grid.place_ship(coords,orientation)
                break
        self.ship_val.pop(0)
```
