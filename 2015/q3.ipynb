```python
import itertools
import numpy as np
def generate_configs(values,n):
    artists = ['A','B','C','D']
    counter = 0
    artworks = []
    for value in values:
        artist = artists[counter]
        for i in range(value):
            artworks.append(artist)
        counter += 1
    permutations = list(itertools.permutations(artworks, len(artworks)))
    unique_permu = list(set(permutations))
    unique_permu.sort()
    
    return unique_permu

permus = generate_configs([2,2,2,2],0)
```
