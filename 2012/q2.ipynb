```python
'''
entering along curve must exit through straight track
lazy point: will set exit from straight track to be the past entrance
flip-flop point: next time, will set exit to be the other track
'''
import string

class Railway:
    def __init__(self):
        pass
        
        
class Point:
    def __init__(self,letter):
        self.flipflop = False
        self.straight = []
        self.curved = []
        self.letter = letter
        self.exit = None
    def list_connections(self):
        print("STRAIGHT:",self.straight)
        for connect in self.straight:
            print(connect.letter)
        print("CURVE:",self.curved)
        for connect in self.curved:
            print(connect.letter)
        
    def pass_through(self,entry):
        if entry in self.curved:
            if self.flipflop == False:
            # If an exit has been set before, follow the exit
                self.exit = entry
                return self.straight[0]
            else:
                return self.straight[0]
        
        if entry in self.straight:
            if self.flipflop == False:
                if self.exit:
                    return self.exit
                else:
                    return self.curved[0]
            else:
                if self.exit:
                    exit = self.exit
                    self.exit = [point for point in self.curved if point != self.exit][0]
                    return exit
                else:
                    exit = self.curved[0]
                    self.exit = [point for point in self.curved if point != self.exit][0]
                    return exit

    def __str__(self):
        return self.letter
```

```python
connections = [[['D'],['E','F']],
 [['C'],['G','H']],
 [['B'],['I','J']],
 [['A'],['K','L']],
 [['A'],['M','N']],
 [['A'],['N','O']],
 [['B'],['O','P']],
 [['B'],['P','Q']],
 [['C'],['Q','R']],
 [['C'],['R','S']],
 [['D'],['S','T']],
 [['D'],['T','M']],
 [['U'],['L','E']],
 [['U'],['E','F']],
 [['V'],['F','G']],
 [['V'],['G','H']],
 [['W'],['H','I']],
 [['W'],['I','J']],
 [['X'],['J','K']],
 [['X'],['K','L']],
 [['V'],['M','N']],
 [['U'],['O','P']],
 [['X'],['Q','R']],
 [['W'],['S','T']]]
```

```python
alphabet_char = string.ascii_uppercase
flipflop_points = ''
flipflop_points = [alphabet_char.index(point) for point in flipflop_points]     
alphabet = []
for i in range(24):
    alphabet.append(Point(alphabet_char[i]))
for point in flipflop_points:
    alphabet[point].flipflop = True
point_dict = {alphabet_char[i]:alphabet[i] for i in range(len(alphabet))}
    
for i in range(len(alphabet)):
    straight = [point_dict[connections[i][0][0]]]
    curved = [point_dict[connections[i][1][0]],point_dict[connections[i][1][1]]]
    alphabet[i].straight = straight
    alphabet[i].curved = curved
    
final_config = ''
track = ('P','V')
iterations = 1000
for i in range(iterations):
    exit = point_dict[track[-1]].pass_through(point_dict[track[0]])
    track = [track[-1],exit.letter]
    print(track)
    final_config += track[0]
print(final_config)
```
