```python
import numpy as np
roman_num_dict = {
    1 : "I",
    5: "V",
    10: "X",
    50: "L",
    100: "C",
    500: "D",
    1000: "M"
}
order = ['M','D','C','L','X','V','I']
def exceptions(num):
    if num == 4:
        return "IV"
    if num == 9:
        return "IX"
    if num == 40:
        return "XL"
    if num == 90:
        return "XC"
    if num == 400:
        return "CD"
    if num == 900:
        return "CM"
    
def num_to_rom(num):
    final_string = ''
    m = num // 1000 
    for i in range(m):
        final_string += 'M'
    num -= 1000*m
    
    if exceptions(num):
        return exceptions(num)
    
    d = num // 500
    for i in range(d):
        final_string += 'D'
    num -= 500*d
    if exceptions(num):
        return exceptions(num)
    c = num // 100
    for i in range(c):
        final_string += 'C'
    num -= 100*c
    if exceptions(num):
        return exceptions(num)
    l = num // 50
    for i in range(l):
        final_string += 'L'
    num -= 50*l
    if exceptions(num):
        return exceptions(num)
    x = num // 10
    for i in range(x):
        final_string += 'X'
    num -= 10*x
    if exceptions(num):
        return exceptions(num)
    v = num // 5
    for i in range(v):
        final_string += 'V'
    num -= 5*v
    if exceptions(num):
        return exceptions(num)
    i = num // 1
    for _ in range(i):
        final_string += 'I'
    num -= 5*i
    if exceptions(num):
        return exceptions(num)
    return final_string
```

```python
def look_and_say(string):
    chars = []
    lengths = []
    target = string[0]
    length =0 
    for i in range(len(string)):
        if string[i] == target:
            length += 1
        else:
            chars.append(target)
            lengths.append(length)
            target = string[i]
            length = 1
    chars.append(target)
    lengths.append(length)
    return chars,lengths

def construct_string(chars,lengths):
    final_string = ''
    for i in range(len(lengths)):
        final_string += num_to_rom(lengths[i])
        final_string += chars[i]
    return final_string
        
        
string = "MMDCCCLXXXVIII"
n = 50
for i in range(n):
    chars,lengths = look_and_say(string)
    string = construct_string(chars,lengths)
string.count("I"),string.count("V")
```

```python
possible_numbers= []
for i in range(1,10000):
    string = num_to_rom(i)
    possible_numbers.append(string)
```

```python
descs = []
for i in range(1,4000):
    string = num_to_rom(i)
    chars,lengths = look_and_say(string)
    descs.append(construct_string(chars,lengths))
```
