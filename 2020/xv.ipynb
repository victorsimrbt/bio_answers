```python
odd_numbers =[number for number in range(1,10000)if number % 2 != 0]
```

```python
def update_numbers(numbers,idx):
    lucky_number = numbers[idx]
    new_numbers = [numbers[i] for i in range(len(numbers)) if not((i+1) % lucky_number == 0)]
    return new_numbers
```

```python
lucky_numbers = []

input_list = odd_numbers
for i in range(1000):
    lucky_numbers.append(input_list[i+1])
    input_list = update_numbers(input_list,i+1)
```
