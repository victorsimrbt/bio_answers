```python
all_connect = []
for entry in perm:
    a = AlphaComplex(8,list(entry))
    connections = 0
    connect = []
    #print(len(a.chosen))
    while len(a.chosen) > 2 and connections < 4:
        for i in range(len(a.alpha)):
            if len(a.chosen) == 2 or connections >= 4:
                break
            if not(a.alpha[i] in a.plan) and a.alpha[i] in a.chosen:
                #print(self.alpha[i],"CONNECTED TO",self.plan[0])
                a.rooms[a.alpha[i]].connections[a.plan[0]] = 0
                a.rooms[a.plan[0]].connections[a.alpha[i]] = 0
                a.chosen.remove(a.alpha[i])
                a.plan.pop(0)
                connections += 1
                connect.append(sorted([a.alpha[i],a.plan[0]]))
                break
    all_connect.append(sorted(connect))
```
