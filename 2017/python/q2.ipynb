```python
import numpy as np

class Grid:
    def __init__(self):
        self.lines = []
        self.grid = np.arange(36).reshape(6,6)
        self.line_own = np.zeros(36*36).reshape(36,36)
    def draw_line_idx(self,a,b,p):
        point_a = self.grid[a[0][0]][a[1][0]]
        point_b = self.grid[b[0][0]][b[1][0]]
        if np.linalg.norm(np.array(a)-np.array(b)) and self.line_own[point_a][point_b] == 0:
            self.lines.append([point_a,point_b])
            self.line_own[point_a][point_b] = p
            self.line_own[point_b][point_a] = p
            return a,b
    def find_squares(self):
        all_squares = []
        for i in range(29):
            if i % 6 == 0 and i != 0:
                pass
            else:
                square_lines = [self.line_own[i][i+1],
                                self.line_own[i+1][i+7],
                                self.line_own[i+7][i+6],
                                self.line_own[i+6][i]]
                all_squares.append(square_lines)
        self.all_squares = all_squares
        return all_squares
    
    def plot_grid(self):
        self.find_squares()
        def string_gen(square_check):
            if square_check == [1,1,1,1]:
                return 'X'
            elif square_check == [-1,-1,-1,-1]:
                return 'O'
            else:
                return '*'
        for _ in range(5):
            a = [string_gen(self.all_squares[i])+' ' if i < 5 else string_gen(self.all_squares[i]) for i in range(5)]
            print(''.join(a))
        
class Player:
    def __init__(self,position,mod,marker):
        self.position = position
        self.mod = mod
        self.marker = marker
    def update_position(self):
        self.position += self.mod
        while self.position > 35:
            self.position -= 36
    def draw_line(self,grid,checker_func):
        index = np.where(grid.grid == self.position)
        print("INDEX",index)
        while True:
            index = np.where(grid.grid == self.position)
            adjacent = checker_func(index)
            for point in adjacent:
                print(point[0],point[1])
                if point[0] < 0 or point[1] < 0 or point[0] >= 6 or point[1] >= 6:
                    pass
                else:
                    point_a = grid.grid[point[0][0]][point[1][0]]
                    point_b = grid.grid[index[0][0]][index[1][0]]
                    if grid.line_own[point_a][point_b] == 0:
                        grid.draw_line_idx(index,point,self.marker)
                        return point_a,point_b
            print("WOAH")
            print(self.position)
            self.position += 1
            while self.position > 35:
                self.position -= 36
```

```python
def player1_check(point):
    return [[point[0]-1,point[1]],[point[0],point[1]+1],[point[0]+1,point[1]],[point[0],point[1]-1]]

def player2_check(point):
    return [[point[0]-1,point[1]],[point[0],point[1]-1],[point[0]+1,point[1]],[point[0],point[1]+1]]
```

```python
game = Grid()
player1 = Player(4-1,10,1)
player2 = Player(14-1,23,-1)

for _ in range(47):
    print("TURN",_)
    player1.update_position()
    print("Player 1 position",player1.position)
    index,point = player1.draw_line(game,player1_check)
    print(index,point)
    print()
    player2.update_position()
    print("Player 2 position",player2.position)
    index,point = player2.draw_line(game,player2_check)
    print(index,point)
    print()
```
