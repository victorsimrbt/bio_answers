```python
colours = ['R','B','G']
def generate_next_row(row):
    pairs = [row[i:i+2] for i in range(len(row)-1)]
    new_values = []
    for pair in pairs:
        if pair[0] == pair[1]:
            new_values.append(pair[0])
        else:
            colour_copy = colours[:]
            colour_copy.remove(pair[0])
            colour_copy.remove(pair[1])
            new_values.append(colour_copy[0])
    return new_values

first_row = input()
print(first_row)

latest_row = first_row
while len(latest_row) != 1:
    latest_row = generate_next_row(latest_row)
print(latest_row)
```
